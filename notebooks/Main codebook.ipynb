```python
import numpy as np
import pandas as pd

data = pd.read_csv('submissions.csv')
```

```python
mh = data[data['outcome'] == 'materialHardship']
ev = data[data['outcome'] == 'eviction']
gpa = data[data['outcome'] == 'gpa']
gr = data[data['outcome'] == 'grit']
jt = data[data['outcome'] == 'jobTraining']
lo = data[data['outcome'] == 'layoff']
```

```python
groups = ev.groupby('account')
```

```python
from sklearn.metrics import roc_auc_score
def CrossEntropyLoss(x):
    filtered = x[x['truth'].notna() & x['prediction'].notna()]
    filtered = filtered[(filtered['prediction'] < 1) & (filtered['prediction'] > 0)]
    indivLoss = filtered['truth']*np.log(filtered['prediction']) + (1-filtered['truth'])*np.log(1 - filtered['prediction'])
    return round(indivLoss.mean(),6)
def r2holdout(x):
    sqdiff = (x['truth']-x['prediction'])**2
    trsqdiff = (x['truth'] - x['ybar_train'])**2
    return round(1-sqdiff.sum()/trsqdiff.sum(),6)
def rocauc(x):
    filtered = x[x['truth'].notna() & x['prediction'].notna()]
    filtered = filtered[(filtered['prediction'] < 1) & (filtered['prediction'] > 0)]
    if not filtered.empty:
        return round(roc_auc_score(filtered['truth'], filtered['prediction']),6)
    else:
        return np.nan
```

```python
cel = ev.groupby('account').apply(CrossEntropyLoss, include_groups=False)
r2h = ev.groupby('account').apply(r2holdout, include_groups=False)
auc = ev.groupby('account').apply(rocauc, include_groups=False)
valid = cel.notna() & r2h.notna() & auc.notna() & (auc != .5)
```

```python
cel_valid = cel[valid]
r2h_valid = r2h[valid]
auc_valid = auc[valid]
```

```python
celRanks = cel_valid.rank(ascending = False, method = 'min')
r2hRanks = r2h_valid.rank(ascending = False, method = 'min')
aucRanks = auc_valid.rank(ascending = False, method = 'min')
rankings = pd.DataFrame({
    'CEL_Rank': celRanks,
    'R2H_Rank': r2hRanks,
    'AUC_Rank': aucRanks
})
```
